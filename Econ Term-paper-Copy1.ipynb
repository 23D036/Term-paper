```python
import pandas as pd

# Load the CSV file
data = pd.read_csv('IHME_GBD_2010_MORTALITY_AGE_SPECIFIC_BY_COUNTRY_1970_2010.csv')

# Group the data by age group, sex, and country
grouped_data = data.groupby(['Age Group', 'Sex', 'Year', 'Country Name'])

# Calculate the maximum and minimum values for each group
max_values = grouped_data['Number of Deaths'].max().reset_index()
min_values = grouped_data['Number of Deaths'].min().reset_index()

# Reshape the maximum values dataframe to tidy format
max_values_tidy = pd.melt(max_values, id_vars=['Age Group', 'Sex', 'Year', 'Country Name'],
                          value_vars=['Number of Deaths'],
                          var_name='Number', value_name='Maximum Value')

# Reshape the minimum values dataframe to tidy format
min_values_tidy = pd.melt(min_values, id_vars=['Age Group', 'Sex', 'Year', 'Country Name'],
                          value_vars=['Number of Deaths'],
                          var_name='Number', value_name='Minimum Value')

# Merge the maximum and minimum tidy dataframes
result_tidy = pd.merge(max_values_tidy, min_values_tidy, on=['Age Group', 'Sex', 'Year','Country Name', 'Number'])

# Print the tidy result for each age group
age_groups = result_tidy['Age Group'].unique()
for age_group in age_groups:
    print(f"Age Group: {age_group}")
    print(result_tidy[result_tidy['Age Group'] == age_group].to_string(index=False))
    print("\n")
```

```python
# Calculate the maximum and minimum values for each group
max_values = grouped_data[['Death Rate Per 100,000']].max().reset_index()
min_values = grouped_data[['Death Rate Per 100,000']].min().reset_index()

# Reshape the maximum values dataframe to tidy format
max_values_tidy = pd.melt(max_values, id_vars=['Age Group', 'Sex', 'Year', 'Country Name'],
                          value_vars=['Death Rate Per 100,000'],
                          var_name='Number', value_name='Maximum Value')

# Reshape the minimum values dataframe to tidy format
min_values_tidy = pd.melt(min_values, id_vars=['Age Group', 'Sex', 'Year', 'Country Name'],
                          value_vars=['Death Rate Per 100,000'],
                          var_name='Number', value_name='Minimum Value')

# Merge the maximum and minimum tidy dataframes
result_tidy = pd.merge(max_values_tidy, min_values_tidy, on=['Age Group', 'Sex', 'Year', 'Country Name', 'Number'])

# Print the result
age_groups = result_tidy['Age Group'].unique()
for age_group in age_groups:
    print(f"Age Group: {age_group}")
    print(result_tidy[result_tidy['Age Group'] == age_group].to_string(index=False))
    print("\n")
```
